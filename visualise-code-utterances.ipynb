## Create radar chart of team 6 (lost, low cohesion) and 22 (lost, high cohesion)

```python
import numpy as np
import matplotlib.pyplot as plt

radar_chart_data = pd.read_csv('/Users/ett506/Documents/proj-league/data/radar_chart_data.csv')

radar_chart_data
```

```python
# select rows whose row values in team_number are equal to team 6
team_6 = radar_chart_data.loc[radar_chart_data['team_number'] == 6]
team_22 = radar_chart_data.loc[radar_chart_data['team_number'] == 22]
```

```python
# get the values across columns as an array
# start at index 1 to ignore the team_number column 

team_6_category_upm = team_6.iloc[:, 1:].values
team_22_category_upm = team_22.iloc[:, 1:].values
```

```python
import plotly.graph_objects as go
import plotly.offline as pyo


categories = ['opinion','obs','command','suggest','thanks','intentions', 'questions', 'answers', 'emotionex',
              'agree', 'humour', 'apology', 'encourage', 'frust', 'disagree', 'nontask', 'misc']

categories = [*categories, categories[0]]

team_6_category_upm = team_6.iloc[:, 1:].values
team_22_category_upm = team_22.iloc[:, 1:].values
team_6_category_upm = [*team_6_category_upm, team_6_category_upm[0]]
team_22_category_upm = [*team_22_category_upm, team_22_category_upm[0]]


fig = go.Figure(
    data=[
        go.Scatterpolar(r=team_6_category_upm, theta=categories, name='Restaurant 1'),
        go.Scatterpolar(r=team_22_category_upm, theta=categories, name='Restaurant 2')
    ],
    layout=go.Layout(
        title=go.layout.Title(text='Restaurant comparison'),
        polar={'radialaxis': {'visible': True}},
        showlegend=True
    )
)

pyo.plot(fig)
```

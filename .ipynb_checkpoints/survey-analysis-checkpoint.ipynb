# Reverse score the HEXA keys

```python
# to reverse score the keys, separate the positive from the negatively keyed

HEXA_positive = ['AGR3', 'EX4', 'EMO5', 'HH6', 'EMO11', 'EX16', 'EMO17', 'HH18', 'EX22', 'EMO23', 'AGR27', 'EMO29', 'AGR33', 'EX34', 'HH36', 'AGR39', 'EX40', 'AGR45', 'EMO47', 'AGR51', 'HH54', 'EX58']

HEXA_negative = ['AGR9R', 'EX10R', 'HH12R', 'AGR15R', 'AGR21R', 'HH24R', 'EX28R', 'HH30R', 'EMO35R', 'EMO41R', 'HH42R', 'EX46R', 'HH48R', 'EX52R', 'EMO53R', 'AGR57R', 'EMO59R', 'HH60R']

len(HEXA_positive)
```

## Function to reverse score HEXA factors

```python
# create a function to reverse score the items
# df = data frame, high = highest score, col = list of columns to reverse score 
def reverse_scoring_hexa(df, high, cols):
    df[cols] = high - df[cols]
                    
    return df

recoded_survey_scores_df = reverse_scoring_hexa(raw_survey_scores_df, 6, HEXA_negative)

recoded_survey_scores_df.to_csv('recoded_survey_scores.csv')
```

# Function to calculate factors

https://hexaco.org/downloads/ScoringKeys_60.pdf

```python
# For factor scores: calculate means across all items in the factor

# creating a new column that represents the mean, for each factor
recoded_survey_scores_df['HH_mean'] = recoded_survey_scores_df[['HH6', 'HH12R', 'HH18', 'HH24R', 'HH30R', 'HH36', 'HH42R', 'HH48R', 'HH54', 'HH60R']].mean(axis=1)
recoded_survey_scores_df['EMO_mean'] = recoded_survey_scores_df[['EMO5', 'EMO11', 'EMO17', 'EMO23', 'EMO29', 'EMO35R', 'EMO41R', 'EMO47', 'EMO59R','EMO53R']].mean(axis=1)
recoded_survey_scores_df['EX_mean'] = recoded_survey_scores_df[['EX4', 'EX10R', 'EX16', 'EX22', 'EX28R', 'EX34', 'EX40', 'EX46R', 'EX52R', 'EX58']].mean(axis=1)
recoded_survey_scores_df['AGR_mean'] = recoded_survey_scores_df[['AGR3', 'AGR9R', 'AGR15R', 'AGR21R', 'AGR27', 'AGR33', 'AGR39', 'AGR45', 'AGR51', 'AGR57R']].mean(axis=1)
recoded_survey_scores_df['RME_total_score'] = recoded_survey_scores_df[['RME1', 'RME2', 'RME3', 'RME4', 'RME5', 'RME6', 'RME7', 'RME8', 'RME9', 'RME10', 'RME11', 'RME12', 
                                                                       'RME13', 'RME14', 'RME15', 'RME16', 'RME17', 'RME18', 'RME19', 'RME20', 'RME21',
                                                                       'RME22', 'RME23', 'RME24', 'RME25', 'RME26', 'RME27', 'RME28', 'RME29', 'RME30', 'RME31', 'RME32',
                                                                       'RME33', 'RME34', 'RME35', 'RME36']].sum(axis=1)


recoded_survey_scores_df.to_csv('with_hexa_means.csv')

recoded_survey_scores_df.head()

# For facet scores: calculate means across all items in the facet. The 60 item test is not expected to have high internal validity for facet-level scores (ignoring this)
```

# Get summary stats of each team using the describe() function

```python
# calculate summary stats for each personality factor per team
HH_summary_stats_df = recoded_survey_scores_df.groupby('team_no')['HH_mean'].describe()
EMO_summary_stats_df = recoded_survey_scores_df.groupby('team_no')['EMO_mean'].describe()
EX_summary_stats_df = recoded_survey_scores_df.groupby('team_no')['EX_mean'].describe()
AGR_summary_stats_df = recoded_survey_scores_df.groupby('team_no')['AGR_mean'].describe()
RME_summary_stats_df = recoded_survey_scores_df.groupby('team_no')['RME_total_score'].describe()
```

```python
HH_summary_stats_df
```

## Rename the column names to identify columns after merging df

```python
# rename the columns 
HH_summary_stats_df = HH_summary_stats_df.rename(columns={"mean": "HH_mean", "std": "HH_std", "min": "HH_min", "25%" : "HH_25_percent", "50%" : "HH_50_percent", "75%" : "HH_75_percent", "max" : "HH_max"})
EMO_summary_stats_df = EMO_summary_stats_df.rename(columns={"mean": "EMO_mean", "std": "EMO_std", "min": "EMO_min", "25%" : "EMO_25_percent", "50%" : "EMO_50_percent", "75%" : "EMO_75_percent", "max" : "EMO_max"})
EX_summary_stats_df = EX_summary_stats_df.rename(columns={"mean": "EX_mean", "std": "EX_std", "min": "EX_min", "25%" : "EX_25_percent", "50%" : "EX_50_percent", "75%" : "EX_75_percent", "max" : "EX_max"})
AGR_summary_stats_df = AGR_summary_stats_df.rename(columns={"mean": "AGR_mean", "std": "AGR_std", "min": "AGR_min", "25%" : "AGR_25_percent", "50%" : "AGR_50_percent", "75%" : "AGR_75_percent", "max" : "AGR_max"})
RME_summary_stats_df = RME_summary_stats_df.rename(columns={"mean": "RME_mean", "std": "RME_std", "min": "RME_min", "25%" : "RME_25_percent", "50%" : "RME_50_percent", "75%" : "RME_75_percent", "max" : "RME_max"})
```

```python
frames = [HH_summary_stats_df, EMO_summary_stats_df, EX_summary_stats_df, AGR_summary_stats_df, RME_summary_stats_df]
pre_test_summary_stats = pd.concat(frames, axis=1)

```

```python
# Adding the match_id and match_outcome

match_info_df = recoded_survey_scores_df.loc[:, ['team_no', 'match_id', 'outcome']]

# drop duplicates so you only have one entry per team
match_info_df = match_info_df.drop_duplicates()
```

## Join match info and survey scores together
## Save it out as a csv

```python
summary_stats = (pd.merge(match_info_df, pre_test_summary_stats, on='team_no'))

# save to csv
# remove index column of df
summary_stats.to_csv('pre_test_summary_stats.csv', index = False)
```

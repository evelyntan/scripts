### function to get team number

```python
def get_team_number_from_xls(transcript):

    split_file_name_into_content = transcript.split("\\")
    #print(split_file_name_into_content)
    team_number_file_name = split_file_name_into_content[6][:-5]
    #print(team_number_file_name)
    split = team_number_file_name.split("-")

    team_number = split[1]
   
    return team_number
```

## re code all files with new coding scheme

```python
import glob
import pandas as pd

# go through the files in the directory
transcripts = glob.glob(r'D:\Projects\UG-league-project\data\coded-transcripts\new-coding-scheme\*.xlsx')

for transcript in transcripts: 
    # get team number
    team_number = get_team_number_from_xls(transcript)
    
    print(team_number)
    
    df = pd.read_excel(transcript)
    
    #extract the rater column values

    rater1_column1 = df.iloc[:, 4]
    rater1_column2 = df.iloc[:, 5]

    rater2_column1 = df.iloc[:, 6]
    rater2_column2 = df.iloc[:, 7]

    rater1_column1_codes = list(rater1_column1.values)
    rater1_column2_codes = list(rater1_column2.values)

    rater2_column1_codes = list(rater2_column1.values)
    rater2_column2_codes = list(rater2_column2.values)
    
    # run the function
    recoded_rater1_1 = recode_new_codes(rater1_column1_codes)
    recoded_rater1_2 = recode_new_codes(rater1_column2_codes)
    recoded_rater2_1 = recode_new_codes(rater2_column1_codes)
    recoded_rater2_2 = recode_new_codes(rater2_column2_codes)
     
    # give new columns
    df['rater1_1'] = recoded_rater1_1
    df['rater1_2'] = recoded_rater1_2
    df['rater2_1'] = recoded_rater2_1
    df['rater2_2'] = recoded_rater2_2
    
    display(df.head())
    
    df.to_excel(r'D:\Projects\UG-league-project\data\coded-transcripts\recoded-transcripts\team-' + team_number + '.xlsx', index = False)
```
